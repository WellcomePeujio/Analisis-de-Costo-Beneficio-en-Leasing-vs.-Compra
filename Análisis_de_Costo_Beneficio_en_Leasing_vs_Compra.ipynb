# Análisis de Costo-Beneficio en Leasing vs. Compra

**Problema**: Una empresa necesita un equipo cuyo costo es de 50,000 dólares. Si lo compra, pagará esa cantidad hoy y tendrá un costo de mantenimiento anual de 5,000 dólares. Si lo alquila, pagará \$12,000 anuales sin costos adicionales. La tasa de descuento es del 8%.

**Preguntas**:

1. Calcula el **costo presente neto (CPN)** para ambas opciones durante 5 años.  
2. ¿Qué opción es más económica?

**Fórmulas Clave**:

1. Para la opción de compra, el costo presente neto (\$( CPN \$)) es:

$$
CPN_{compra} = C_{inicial} + \sum_{t=1}^{5} \frac{C_{mantenimiento}}{(1 + r)^t}
$$

2. Para la opción de alquiler, el costo presente neto (\$( CPN \$)) es:

$$
CPN_{leasing} = \sum_{t=1}^{5} \frac{C_{anual}}{(1 + r)^t}
$$

Donde:
- \$( C_{inicial} \$): Costo inicial de compra (\$50,000).  
- \$( C_{mantenimiento} \$): Costo anual de mantenimiento (\$5,000).  
- \$( C_{anual} \$): Costo anual de alquiler (\$12,000).  
- \$( r \$): Tasa de descuento (8% o 0.08).  
- \$( t \$): Años (1 a 5).


# Resolución Matemática: Análisis de Costo-Beneficio en Leasing vs. Compra

## Paso 1: Fórmulas para el Costo Presente Neto

1. **Opción de Compra**:  
El costo presente neto (\$( CPN_{compra} \$)) se calcula como:

$$
CPN_{compra} = C_{inicial} + \sum_{t=1}^{5} \frac{C_{mantenimiento}}{(1 + r)^t}
$$

2. **Opción de Leasing**:  
El costo presente neto (\$( CPN_{leasing} \$)) se calcula como:

$$
CPN_{leasing} = \sum_{t=1}^{5} \frac{C_{anual}}{(1 + r)^t}
$$

---

## Paso 2: Sustituir los Valores Dados

- \$( C_{inicial} = 50,000 \$)  
- \$( C_{mantenimiento} = 5,000 \$)  
- \$( C_{anual} = 12,000 \$)  
- \$( r = 0.08 \$)  

---

### Cálculo para la Opción de Compra

Sustituyendo en la fórmula:

$$
CPN_{compra} = 50,000 + \sum_{t=1}^{5} \frac{5,000}{(1 + 0.08)^t}
$$

Calculamos cada término de la serie:

1. Para \$( t = 1 \$):

   $$
   \frac{5,000}{(1 + 0.08)^1} = \frac{5,000}{1.08} \approx 4,629.63
   $$

2. Para \$( t = 2 \$):

   $$
   \frac{5,000}{(1 + 0.08)^2} = \frac{5,000}{1.1664} \approx 4,287.79
   $$

3. Para \$( t = 3 \$):

   $$
   \frac{5,000}{(1 + 0.08)^3} = \frac{5,000}{1.2597} \approx 3,968.49
   $$

4. Para \$( t = 4 \$):

   $$
   \frac{5,000}{(1 + 0.08)^4} = \frac{5,000}{1.3605} \approx 3,676.62
   $$

5. Para \$( t = 5 \$):

   $$
   \frac{5,000}{(1 + 0.08)^5} = \frac{5,000}{1.4693} \approx 3,408.17
   $$

Sumamos los costos de mantenimiento:

$$
\text{Costo mantenimiento total} = 4,629.63 + 4,287.79 + 3,968.49 + 3,676.62 + 3,408.17 \approx 19,970.70
$$

Finalmente, el costo presente neto de compra es:

$$
CPN_{compra} = 50,000 + 19,970.70 = 69,970.70
$$

---

### Cálculo para la Opción de Leasing

Sustituyendo en la fórmula:

$$
CPN_{leasing} = \sum_{t=1}^{5} \frac{12,000}{(1 + 0.08)^t}
$$

Calculamos cada término de la serie:

1. Para \$( t = 1 \$):

   $$
   \frac{12,000}{(1 + 0.08)^1} = \frac{12,000}{1.08} \approx 11,111.11
   $$

2. Para \$( t = 2 \$):

   $$
   \frac{12,000}{(1 + 0.08)^2} = \frac{12,000}{1.1664} \approx 10,293.49
   $$

3. Para \$( t = 3 \$):

   $$
   \frac{12,000}{(1 + 0.08)^3} = \frac{12,000}{1.2597} \approx 9,524.77
   $$

4. Para \$( t = 4 \$):

   $$
   \frac{12,000}{(1 + 0.08)^4} = \frac{12,000}{1.3605} \approx 8,802.90
   $$

5. Para \$( t = 5 \$):

   $$
   \frac{12,000}{(1 + 0.08)^5} = \frac{12,000}{1.4693} \approx 8,125.96
   $$

Sumamos los costos de leasing:

$$
CPN_{leasing} = 11,111.11 + 10,293.49 + 9,524.77 + 8,802.90 + 8,125.96 \approx 47,858.23
$$

---

## Conclusión

1. **Costo Presente Neto de Compra**:

   $$ CPN_{compra} = 69,970.70 \, \text{dólares} $$  

2. **Costo Presente Neto de Leasing**:  
   $$ CPN_{leasing} = 47,858.23 \, \text{dólares} $$  

3. **Opción más económica**: Leasing es la opción más económica, con un ahorro aproximado de:

$$
69,970.70 - 47,858.23 = 22,112.47 \, \text{dólares}
$$


```python
import numpy as np
import matplotlib.pyplot as plt

# Parámetros del problema
costo_inicial = 50000  # Costo inicial de compra ($)
costo_mantenimiento = 5000  # Costo anual de mantenimiento ($)
costo_leasing = 12000  # Costo anual de leasing ($)
tasa_descuento = 0.08  # Tasa de descuento
anios = np.arange(1, 6)  # Años: 1 a 5

# Calcular el costo presente neto (CPN) para compra
cpn_compra = costo_inicial + sum(costo_mantenimiento / (1 + tasa_descuento)**t for t in anios)

# Calcular el costo presente neto (CPN) para leasing
cpn_leasing = sum(costo_leasing / (1 + tasa_descuento)**t for t in anios)

# Crear arrays para la gráfica
costos_mantenimiento = [costo_mantenimiento / (1 + tasa_descuento)**t for t in anios]
costos_leasing = [costo_leasing / (1 + tasa_descuento)**t for t in anios]

# Graficar
plt.figure(figsize=(10, 6))

# Barras de costos anuales
plt.bar(anios - 0.2, costos_mantenimiento, width=0.4, label="Mantenimiento (Compra)", color="blue", alpha=0.7)
plt.bar(anios + 0.2, costos_leasing, width=0.4, label="Costo Anual (Leasing)", color="green", alpha=0.7)

# Etiquetas y leyendas
plt.title("Evolución de Costos: Compra vs Leasing")
plt.xlabel("Año")
plt.ylabel("Costo Presente ($)")
plt.xticks(anios)
plt.axhline(0, color="black", linewidth=0.8, linestyle="--")
plt.legend()
plt.grid(axis='y', linestyle="--", alpha=0.7)

# Mostrar gráfica
plt.show()

# Resultados
{
    "CPN Compra": round(cpn_compra, 2),
    "CPN Leasing": round(cpn_leasing, 2),
    "Diferencia de Costos": round(cpn_compra - cpn_leasing, 2)
}
```
